Import the dataset

```python
full_data = pd.read_csv('/content/titanic.csv')
```

```python
full_data.shape
```

```python
full_data.head()
```

2 Distribution Plots

```python
# 2.1 distplot
sns.histplot(full_data['Parch'], kde = False)
plt.show()
```

```python
# 2.2 kdeplot
sns.distplot(full_data['Age'], hist = False)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.distplot(full_data['Age'])

plt.show()
```

Relational plots

```python
sns.relplot(x = 'Age', y = 'Fare', col = 'Pclass', hue = 'Sex', style = 'Sex', kind = 'line', data = full_data)
plt.show()
```

Scatterplot

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x = 'Age', y = 'Fare', hue = 'Sex', data = full_data)
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.lineplot(x = 'Age', y = 'Fare', hue = 'Sex', data = full_data)
plt.show()
```

Catogerical Plot

```python
# 4.1 barplot

plt.figure(figsize=(8, 8))  # Create a figure with specified size
sns.barplot(x='Sex', y='Survived', hue='Pclass', data=full_data)  # Create the bar plot
plt.show()
```

```python
# 4.2 stripplot
plt.figure(figsize=(8, 8))  # Create a figure with specified size
sns.stripplot(x='Sex', y='Age', hue='Sex', data=full_data)  # Create the strip plot
plt.show()
```

```python
# swarmplot
# 4.3
plt.figure(figsize=(8, 8))
sns.swarmplot(x='Sex', y='Age', hue='Sex', data=full_data)
plt.show()
```

```python
# 4.5
# Boxplot

plt.figure(figsize = (8,8))
sns.boxplot(x = 'Survived', y = 'Age', data = full_data)
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.boxplot(x = 'Sex', y = 'Age', data = full_data)
plt.show()
```

```python
# violinplot
sns.violinplot(x = 'Survived', y = 'Age',hue = 'Sex', data = full_data)

plt.show()
```

```python
# countplot
sns.countplot(x = 'Survived', data = full_data, palette = 'Blues')
plt.show()
```

```python
# pointplot

plt.subplots(figsize = (8,8))
sns.pointplot(x = 'Sex', y = 'Survived', hue = 'Pclass', data = full_data)
plt.show()
```

Regression plot

```python
# 5.1 implot
sns.lmplot(x = 'Age', y = 'Fare', data = full_data)
plt.show()
```

```python
# regplot
plt.subplots(figsize = (10,10))
sns.regplot(x = 'Age', y = 'Fare', data = full_data)
plt.show()
```

```python
plt.subplots(figsize = (10,10))
sns.regplot(x = 'Fare', y = 'Fare', data = full_data)
plt.show()
```

```python
# Matrix plot

# 1 Heatmap

plt.subplots(figsize = (10,10))
sns.heatmap(full_data.corr(),cmap = 'YIGnBu', annot = True, fmt = ".2f")

plt.show()
```

```python
# Facetgrid.map
g = sns.FacetGrid(full_data, col = 'Survived', row = 'Pclass')

g.map(plt.hist, 'Age')
g.add_legend()
plt.show()
```

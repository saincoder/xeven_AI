# ***Text Summarization***

we can summarize the text using the hugging face pipline for that we need to pass the pipline the parameers as summarization

```python
import transformers
from transformers import pipeline

# Text to be summarize
input_text = """
Hugging Face is a company specializing in Natural Language Processing (NLP) and machine learning technologies. They are best known for their open-source library, Transformers, which provides a wide range of pre-trained models for tasks like text classification, translation, and question answering. These models are built on architectures like BERT, GPT, and T5.

In addition to Transformers, Hugging Face offers tools for managing machine learning models, such as the `datasets` library for working with datasets and `tokenizers` for efficient tokenization. They also provide a platform called Hugging Face Hub where users can share and discover models and datasets. Their goal is to democratize AI and make state-of-the-art NLP technology accessible to everyone.
"""
print(input_text)
```

```python
summarizer = pipeline("summarization")
summarizer(f"Text Summarization: {input_text}", min_length=5, max_length = 30)
```

# ***Stable Diffusion pipeline***

```python
!pip install diffusers
```

```python
import transformers
from transformers import pipeline
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
promt = "A image of an astronaut riding a horse"
image = pipe(promt).images[0]
image
```

```python
# to save image
image.save("astronaut_rides_horse.png")
```
